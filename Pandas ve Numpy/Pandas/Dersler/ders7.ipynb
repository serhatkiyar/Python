# Join birleştirme işlemleri

```python
import numpy as np
import pandas as pd
```

```python
m = np.random.randint(1, 30, (5, 3))
df1 = pd.DataFrame(m, columns=['Column1', 'Column2', 'Column3'])
df1
```

```python
df2 = df1 + 9
df2
```

# pd.concat(  [ ]  )

```python
pd.concat([df1, df2])
```

**concat metodunun varsayılan axis değeri 0 olduğu için index bazında birleştirme yapar.**

```python
pd.concat([df1, df2], axis=1)
```

**______________________________________________________________________________________________________**

**Concat fonksiyonunu kullanırken index problemi. Yukarıda gördüğümüz gibi index bazında birleştirme yaparken dataframlerin indexleri alt alta koyuldu. ignore_index argümanı ile oluşan dataframe i 0 dan indexledik.**

```python
pd.concat([df1, df2], ignore_index=True)
```

**df2' i dataframe' inin 'Column3' değişkenini değiştirdik. Şimdi tekrar df1 ve df2 dataframeleri üzerine concat metodunu kullanırsak. Ne olucağına bakalım. (Farklı Columnlar var iken index bazında birleştirme.)**

```python
pd.concat([df1, df2])
```

**görüldüğü üzere df1' in 'Column13' değişkeni için bir değer verisi olmadığı için NaN döndü. Bu durum df2' de ise 'Column3' de gerçekleşiyor.**

**______________________________________________________________________________________________________**


**Peki NaN değerleri olan Columnları yani ortak olmayan columnlar olmadan nasıl veriyi alırız. join=**

```python
pd.concat([df1, df2], join='inner')
```
