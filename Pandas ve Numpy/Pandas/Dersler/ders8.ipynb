# İleri Birleştirme İşlemleri

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'Calisanlar' : ['Serhat', 'Umut', 'Muhammed', 'Efe'],
    'Gruplar' : ['Öğretmen', 'Mühendislik', 'Mühendislik', 'Doktor']
}) 
df1
```

```python
df2 = pd.DataFrame({
    'Calisanlar' : ['Serhat', 'Umut', 'Muhammed', 'Efe'],
    'Giris' : ['2010', '2009', '2014', '2017']
})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on = 'Calisanlar')
```
